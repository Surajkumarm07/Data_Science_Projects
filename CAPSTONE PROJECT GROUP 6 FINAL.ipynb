## 3.10  Statistical Test
##  Quantitative Vs Target

Kruskal-Wallis test, 
which is a nonparametric statistical test used to compare the median ranks of three or more independent groups. 
It is often used as an alternative to the parametric one-way analysis of variance (ANOVA) test 
when the assumptions of normality and equal variance are not met.

If p values less than 0.05, we can say categorical groups are different

```python
# Shapiro Wilke Normality Test to Test Normality

# Hypothesis:

# Ho : Null Hypothesis
# Ha : Alternate Hypothesis

# Ho : Data is normal
# Ha : Data is not normal

# Non-Parametric Test to Test Dependency

# Hypothesis:

# Ho : Population medians are equal
# Ha : Population medians are not all equal

for i in num_col:
    print(i,':\n\n',stats.shapiro(data[i]))
    data_group1 = data[data['Credit_Score']=='Poor'][i]
    data_group2 = data[data['Credit_Score']=='Standard'][i]
    data_group3 = data[data['Credit_Score']=='Good'][i]
    print(i,':',stats.kruskal(data_group1, data_group2, data_group3),'\n\n')
```

Interpetation:-Since from the above results we conclude that p_value is less than the significance level so, we reject null hypothesis and concluding that Data is not Normal and Credit Score for each category is dependent to each other for given independent variables.

## Category Vs Target

```python
# Hypothesis

# Ho : Null Hypothesis
# Ha : Alternate Hypothesis

# Ho : Category vs Target are Independent / Category has no impact on Target
# Ha : Category vs Target are Dependent / Category has Impact on Target

for i in cat_col:
    stat,pval,ddof,exp_val = stats.chi2_contingency(pd.crosstab(data[i],data['Credit_Score']))
    print(i,':\nP_Value',pval,'\n')
```

Interpetation:- Since from the above results we conclude that p_value is less than the significance level so, we reject null hypothesis and concluding that Credit Score for each category is dependent.

# 4.0  Model Building

Creating a user defined function for - Classification report, Confusion matrix, ROC-AUC curve

```python
def get_test_report(model):
    
    test_pred = model.predict(X_test)

    # return the classification report for test data
    return (classification_report(y_test,test_pred))
```

```python
def plot_confusion_matrix(model):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    conf_matrix = pd.DataFrame(data=cm, 
                           columns=['Predicted: 0', 'Predicted: 1', 'Predicted: 2'], 
                           index=['Actual: 0', 'Actual: 1', 'Actual: 2'])
    sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = 'Blues', cbar = False, 
                linewidths = 0.1, annot_kws = {'size':20})

    plt.xticks(fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.show()
```

```python
def plot_feature_importance(model, feature_names):
    feature_importance = pd.DataFrame({'Feature': feature_names,
                                        'Importance': model.feature_importances_})
    feature_importance = feature_importance.sort_values('Importance', ascending=False)

    # plot the bar chart
    plt.figure(figsize=(10, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importance)

    # add plot and axes labels
    plt.title('Feature Importance', fontsize=15)
    plt.xlabel('Importance', fontsize=12)
    plt.ylabel('Features', fontsize=12)

    plt.show()
```

```python
clfs = {
    "Logreg":LogisticRegression(),
    "KNN":KNeighborsClassifier(),
    "Navie Bayes":GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost":AdaBoostClassifier(),
    "GradientBoosting":GradientBoostingClassifier(),
    "XGboost":XGBClassifier()
}

models_report = pd.DataFrame(columns=["Model","Accuracy","Recall","Precision","F1"])


for clf,clf_name in list(zip(clfs.values(), clfs.keys())):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print("Fitting the model.....",clf_name)
    t = pd.Series({
        "Model":clf_name,
        "Accuracy":accuracy_score(y_test, y_pred),
        "Recall":recall_score(y_test, y_pred,average=None),
        "Precision":precision_score(y_test, y_pred,average='macro'),
        "F1":f1_score(y_test, y_pred,average='weighted')
        
    })
    models_report = models_report.append(t, ignore_index=True)   
models_report = models_report.sort_values(by="F1",ascending= False)
models_report
```
